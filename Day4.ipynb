```python
# Numpy (Python에서 과학 계산을 위한 필수 라이브러리, 대규모 다차원 배열과 행렬 연산을 효율적으로)
import numpy as np
arr = np.array([1, 2, 3]) # 리스트를 Numpy 배열로 변환
arr = np.arange(0, 10, 2) # 주어진 범위 내에서 일정한 간격의 값을 가지는 배열 생성(0~9까지 2씩 증가)
                          # arrange(시작값, 종료값, 건너뛰기)
arr = np.random.rand(3, 3) # 0과 1 사이의 균일 분포를 따르는 난수 생성(3, 3일 경우 2차원 배열 3행 3열)
                           # np.ramdom : 난수생성, rand: 0~1난수
arr = arr.reshape(1, 9) # 배열의 shape 변경(1행 9열로 변경)(ex)rand로 3행 3열로 만들고 1행 9열로 변경)
arr
np.sum([1, 2, 3]) # 배열 요소의 합
np.mean([1, 2, 3]) # 배열 요소의 평균
np.max([175, 177, 179, 181, 183]) # 배열의 최댓값
np.std([175, 177, 179, 181, 183]) # 배열 요소의 표준편차 계산
                  #표준편차 : 값 - 평균 = ? > ? 제곱 = * > * / 인원수 > 루트 
```
