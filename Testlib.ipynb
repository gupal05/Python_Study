```python
# 100개의 난수를 포함하는 배열을 생성하고, 그 배열의 합, 평균, 최댓값, 최솟값을 구해 보세요.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

arr = np.random.rand(100)
print(np.sum(arr))
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
```

```python
# 아래의 데이터를 사용하여 Pandas 데이터프레임을 만들고, Age와 Salary 컬럼에 대한 평균과 중앙값을 구하세요.
data = {
    'Name' : ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age' : [25, 30, 35, 40, 45],
    'Salary' : [50000, 60000, 70000, 80000, 90000]
}
df = pd.DataFrame(data)
age_avg = df['Age'].mean() # 평균값
age_median = df['Age'].median() # 중앙값
salary_avg = df['Salary'].mean()
salary_median = df['Salary'].median()
print(age_avg)
print(age_median)
print(salary_avg)
print(salary_median)

#위 데이터를 기반으로, Age가 30 이상인 사람들의 Salary 평균을 구하세요.
a = df[df['Age'] >= 30]
print(a['Salary'].mean())
```

```python
# NumPy를 사용하여 0부터 2π까지의 범위에서 100개의 값을 생성한 후, 그 값에 대한
# 함수와 코사인 함수를 같은 그래프에 그려보세요. 그래프에 범례(legend)를 추가하고
# X축과 Y축에 라벨을 붙이세요.
# 추가 문제 : 위의 사인 함수 그래프에 히스토그램을 추가하여 두 함수의 분포를 비교해 보세요.

x = np.linspace(0, 2*np.pi, 100) # 0 ~ 2π까지의 범위에서 100개의 값 생성
y_sin = np.sin(x) # 사인
y_cos = np.cos(x) # 코사인
plt.plot(x, y_sin, label='Sin(x)', color='blue') # 사인함수 그래프 그리기
plt.plot(x, y_cos, label='Cos(x)', color='red') # 코사인 함수 그래프 그리기
# 히스토그램 그리기
plt.hist(y_sin, bins=20, alpha=0.5, label='Histogram of Sin(x)', color='blue', orientation='vertical')
plt.hist(y_cos, bins=20, alpha=0.5, label='Histogram of Cos(x)', color='red', orientation='vertical')
plt.legend() # 범례 추가
plt.xlabel('x') 
plt.ylabel('y')
plt.show()
```
